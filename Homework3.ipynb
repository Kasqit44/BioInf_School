```python
# Task 1
set_1 = set()
set_1.add('Happy')
set_1.add(4)
set_1.add(17.3)

set_2 = {'I am', '123', 'Happy', 4}

print('Union:', set_1 | set_2)
print('Intersection:', set_1 & set_2)
print('Difference:', set_1 - set_2)
print('Symmetric difference', set_1 ^ set_2, '\n\n')

set_3 = {4, 17.3}

print('Is set_3 subset of set_1? ', '\n', set_3.issubset(set_1))
print('Is set_3 subset of set_2? ', '\n', set_3.issubset(set_2))
```

```python
# Task 2

my_dict = {}
my_dict_zz = {}
my_dict['S'] = 12
my_dict['Y'] = 15
my_dict['Z'] = 10101
print('here is my_dict', my_dict, '\n\n')

line = 'q w e r t y u i o p'
position = 1
for letter in line.split():
    my_dict_zz[letter] = position
    position += 1
print('here is my_dict_zz', my_dict_zz, '\n\n')
    
remove = [key for key, value in my_dict_zz.items() if value > 6]
for key in remove:
    del my_dict_zz[key]
        

print(my_dict_zz, '\n\n')

for key, value in my_dict_zz.items():
    if value % 2 == 0:
        my_dict_zz[key] = 'even'
    else: 
        my_dict_zz[key] = 'odd'
print(my_dict_zz)

    
```

```python
# Task 9
DNA = input('write down your DNA sequence: ')

def rev_comp_1 (DNA):
    dna_1 = DNA.upper()
    comp = { 'A' : 'T',
             'T' : 'A',
             'G' : 'C',
             'C' : 'G'}
    list_comp = []
    for i in dna_1:
        i = comp[i]
        list_comp.append(i)
    cDNA = ''.join(list_comp[::-1])
    print(cDNA)



def rev_comp_2(DNA):
    dna_1 = DNA.upper()
    cDNA_list = list(dna_1)
    for nucleotide_position in range(len(cDNA_list)):
        if cDNA_list[nucleotide_position] == 'A':
            cDNA_list[nucleotide_position] = 'T'
        elif cDNA_list[nucleotide_position] == 'T':
            cDNA_list[nucleotide_position] = 'A'
        elif cDNA_list[nucleotide_position] == 'G':
            cDNA_list[nucleotide_position] = 'C'
        elif cDNA_list[nucleotide_position] == 'C':
            cDNA_list[nucleotide_position] = 'G'
    print(''.join(cDNA_list[::-1]))
    
    

def rev_comp_3(DNA):
    dna_1 = DNA.upper()
    dna_2 = dna_1.replace('A', '0')
    dna_2_1 = dna_2.replace('T', '1')
    dna_2_2 = dna_2_1.replace('G', '2')
    dna_2_3 = dna_2_2.replace('C', '3')
    list_c = list(dna_2_3)
    list_cc = []
    for i in list_c:
        if i == '0':
            list_cc.append('T')
        elif i == '1':
            list_cc.append('A')
        elif i == '2':
            list_cc.append('C')
        elif i == '3':
            list_cc.append('G')
    print(''.join(list_cc[::-1]))
    
    

def rev_comp_4(DNA):
    dna_1 = DNA.upper()
    list_comp = []
    for nucleotide in dna_1:
        if nucleotide == 'A':
            list_comp.append('T')
        elif nucleotide == 'T':
            list_comp.append('A')
        elif nucleotide == 'C':
            list_comp.append('G')
        elif nucleotide == 'G':
            list_comp.append('C')
    print(''.join(list_comp[::-1]))

    

def rev_comp_5(DNA):
    dna_1 = DNA.upper()
    dna_2 = dna_1.translate(str.maketrans('ATGC', 'TACG'))
    cDNA = dna_2[::-1]
    print(cDNA)
    
rev_comp_1(DNA)
rev_comp_2(DNA)
rev_comp_3(DNA)
rev_comp_4(DNA)
rev_comp_5(DNA)
    
```
