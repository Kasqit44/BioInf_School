```python
#задание 2
'''
а) формула для нахождения площади круга
б) формула для вычисления кинетической энергии тела
в) формула для нахождения среднего выборки
''' 

#формула А
def sqr(r):
    S = 3.14 * r * r
    print ('радиус круга равен', r)
    print ('площадь круга равна', S)
sqr(4)
sqr(4.2)
sqr(1.1)

print ()

#формула Б
def kin(m, V):
    kin_e = m * V * V / 2
    print (f'Если тело массой {m} кг двигалось со скоростью {V} м/с, то его кинетическая энергия равна {kin_e} Дж')
kin(5, 2)
kin(5.7, 2.1)
kin(1.4, 2.99)

print('')

#формула В
def _mean_(xx):
    summ = 0
    for i in xx:
        summ += i
    mean_xx = summ / len(xx)
    print ('Среднее данной выборки равно ', mean_xx)

_mean_([2, 3, 5])
_mean_([1.1, 4.9, 11, 1.07])
_mean_([2, 2, 2, 2])
```

```python
#задание 3
print('Not')
def _not_(a):
    if a:
        print(f'a = {a} is False')
    else:
        print(f'a = {a} is True')
_not_(True)
_not_(False)
print()
print('Or')
def _or_(a, b):
    if a or b:
        print(f'a = {a} and b = {b} are True')
    else:
        print(f'a = {a} and b = {b} are False')
_or_(False, False)
_or_(True, False)
_or_(False, True)
_or_(True, True)

print()
print ('Xor')
def _xor_(a, b):
    if (a or b) and not (a and b ):
        print(f'a = {a} and b = {b} are True')
    else:
        print(f'a = {a} and b = {b} are False')
_xor_(False, False)
_xor_(True, False)
_xor_(False, True)
_xor_(True, True)

print()
print('Nor')
def _nor_(a, b):
    if not a and not b:
        print(f'a = {a} and b = {b} are True')
    else: 
        print(f'a = {a} and b = {b} are False')

_nor_(False, False)
_nor_(True, False)
_nor_(False, True)
_nor_(True, True)

print()
print('And')
def _and_ (a, b):
    if a and b:
        print(f'a = {a} and b = {b} are True')
    else:
        print(f'a = {a} and b = {b} are False')
_and_(False, False)
_and_(True, False)
_and_(False, True)
_and_(True, True)

print()
print('Nand')
def _nand_(a, b):
    if a and b:
        print(f'a = {a} and b = {b} are False')
    else:
        print(f'a = {a} and b = {b} are True')
_nand_(False, False)
_nand_(True, False)
_nand_(False, True)
_nand_(True, True)
```
