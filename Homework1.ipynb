```python
#задание 2
'''
а) формула для нахождения площади круга
б) формула для вычисления кинетической энергии тела
в) формула для нахождения среднего выборки
''' 

#формула А
def sqr(r):
    S = 3.14*r*r
    print ('радиус круга равен', r)
    print ('площадь круга равна', S)
sqr(4)
sqr(4.2)
sqr(1.1)

print ('')

#формула Б
def kin(m,V):
    kin = m*V*V/2
    print (f'Если тело массой {m} кг двигалось со скоростью {V} м/с, то его кинетическая энергия равна {kin} Дж')
kin(5,2)
kin(5.7,2.1)
kin(1.4,2.99)

print('')

#формула В
def mean(xx):
    sum = 0
    for i in xx:
        sum+=i
    mean = sum/len(xx)
    print ('Среднее данной выборки равно ', mean)

mean([2,3,5])
mean([1.1, 4.9, 11, 1.07])
mean([2,2,2,2])
```

```python
#задание 3
print ('Not')
def Not(a):
    if a == 1:
        print (0)
    else:
        print (1)
Not(1)
Not(0)
print('')
print ('Or')
def Or(a,b):
    if a==0 and b==0:
        print (0)
    else:
        print (1)
Or(0,0)
Or(1,0)
Or(0,1)
Or(1,1)

print('')
print ('Xor')
def Xor(a,b):
    if (a!=0 or b!=0) and a+b!=2:
        print (1)
    else:
        print (0)
Xor(0,0)
Xor(1,0)
Xor(0,1)
Xor(1,1)

print('')
print ('Nor')
def Nor(a,b):
    if not a and not b:
        print(1)
    else: 
        print(0)

Nor(0,0)
Nor(1,0)
Nor(0,1)
Nor(1,1)

print ('')
print ('And')
def And (a,b):
    if a and b:
        print(1)
    else:
        print (0)
And(0,0)
And(1,0)
And(0,1)
And(1,1)

print ('')
print ('Nand')
def Nand(a,b):
    if a and b:
        print (0)
    else:
        print (1)
Nand(0,0)
Nand(1,0)
Nand(0,1)
Nand(1,1)
```
